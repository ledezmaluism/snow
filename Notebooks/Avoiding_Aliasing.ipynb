## How to avoid aliasing when multiplying two signals

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
from numpy.fft import fft, ifft, fftshift, fftfreq
from scipy.constants import pi, c, epsilon_0
```

```python
NFFT = 2**11
t = np.linspace(-1e-12, 1e-12, NFFT)
dt = t[1] - t[0]
f = fftfreq(NFFT, dt)

#Noise floor
noise = 1e-10*np.random.normal(size=NFFT)

#Pulse creation
f0_ff = 200e12
f_ref = 400e12
delta_omega = 2*pi*(f0_ff - f_ref)
tau = 100e-15
x = 1*np.exp(-2*np.log(2)*(t/tau)**2)*np.exp(1j*delta_omega*t)
x = x + noise
X = fft(x)

plt.rcParams['figure.figsize'] = [15, 4]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(t*1e12, np.abs(x)); ax1.set_xlabel('Time (ps)'); ax1.set_xlim([-1,1])
ax2.plot(fftshift(f+f_ref)*1e-12, fftshift(np.abs(X))); ax2.set_xlabel('Frequency (THz)');
```

If now we just square the pulse we should get a second harmonic.

```python
x = x + x**2*np.exp(1j*2*pi*f_ref*t)
X = fft(x)

plt.rcParams['figure.figsize'] = [15, 4]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(t*1e12, np.abs(x)); ax1.set_xlabel('Time (ps)'); ax1.set_xlim([-1,1])
ax2.plot(fftshift(f+f_ref)*1e-12, fftshift(np.abs(X))); ax2.set_xlabel('Frequency (THz)');
```

## How can fix this?

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
from numpy.fft import fft, ifft, fftshift, fftfreq
from scipy.constants import pi, c, epsilon_0
```

```python
NFFT = 2**10
t = np.linspace(-2e-12, 2e-12, NFFT)
dt = t[1] - t[0]
f = fftfreq(NFFT, dt)

#Noise floor
noise = 1e-10*np.random.normal(size=NFFT)

#Pulse creation
f0_ff = 200e12
f_ref = 300e12
delta_omega = 2*pi*(f0_ff - f_ref)
tau = 100e-15
x = 1*np.exp(-2*np.log(2)*(t/tau)**2)*np.exp(1j*delta_omega*t)
x = x + noise
X = fft(x)

plt.rcParams['figure.figsize'] = [15, 4]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(t*1e12, np.abs(x)); ax1.set_xlabel('Time (ps)'); ax1.set_xlim([-1,1])
ax2.plot(fftshift(f+f_ref)*1e-12, fftshift(np.abs(X))); ax2.set_xlabel('Frequency (THz)');
```

```python
k = 0.1

Nup = 5
y = scipy.signal.resample(x, Nup*NFFT) #upsampled signal
ty = np.linspace(t[0], t[-1], Nup*NFFT)
dty = ty[1] - ty[0]
fy = fftfreq(Nup*NFFT, dty)

Y = fft(y)

plt.rcParams['figure.figsize'] = [15, 4]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(ty*1e12, np.abs(y)); ax1.set_xlabel('Time (ps)'); ax1.set_xlim([-1,1])
ax2.plot(fftshift(fy+f_ref)*1e-12, fftshift(np.abs(Y))); ax2.set_xlabel('Frequency (THz)');

y = y + k*y**2*np.exp(1j*2*pi*f_ref*ty)
Y = fft(y)

plt.rcParams['figure.figsize'] = [15, 4]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(ty*1e12, np.abs(y)); ax1.set_xlabel('Time (ps)'); ax1.set_xlim([-1,1])
ax2.plot(fftshift(fy+f_ref)*1e-12, fftshift(np.abs(Y))); ax2.set_xlabel('Frequency (THz)');

x = scipy.signal.resample(y, NFFT) #downsampled signal
X = fft(x)

plt.rcParams['figure.figsize'] = [15, 4]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(t*1e12, np.abs(x)); ax1.set_xlabel('Time (ps)'); ax1.set_xlim([-1,1])
ax2.plot(fftshift(f+f_ref)*1e-12, fftshift(np.abs(X))); ax2.set_xlabel('Frequency (THz)');
```

```python
NFFT = 2**12
t = np.linspace(-1e-12, 1e-12, NFFT)
dt = t[1] - t[0]
f = fftfreq(NFFT, dt)

#Noise floor
noise = 1e-10*np.random.normal(size=NFFT)

#Pulse creation
f0_ff = 200e12
f_ref = 400e12
delta_omega = 2*pi*(f0_ff - f_ref)
tau = 100e-15
x = 1*np.exp(-2*np.log(2)*(t/tau)**2)*np.exp(1j*delta_omega*t)
x = x + noise
X = fft(x)

plt.rcParams['figure.figsize'] = [15, 4]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(t*1e12, np.abs(x)); ax1.set_xlabel('Time (ps)'); ax1.set_xlim([-1,1])
ax2.plot(fftshift(f+f_ref)*1e-12, fftshift(np.abs(X))); ax2.set_xlabel('Frequency (THz)');
```

```python
x = x - 1j*x**2*np.exp(1j*2*pi*f_ref*t) - x*np.conj(x)*np.exp(-1j*2*pi*f_ref*t)
X = fft(x)

plt.rcParams['figure.figsize'] = [15, 4]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(t*1e12, np.abs(x)); ax1.set_xlabel('Time (ps)'); ax1.set_xlim([-1,1])
ax2.plot(fftshift(f+f_ref)*1e-12, fftshift(np.abs(X))); ax2.set_xlabel('Frequency (THz)');
```
